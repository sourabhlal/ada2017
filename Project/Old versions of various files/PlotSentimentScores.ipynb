```python
import pandas as pd
import gzip
from tqdm import tqdm

from datetime import datetime
from datetime import date, time
from dateutil.parser import parse
import numpy as np
import pandas as pd


from bokeh.plotting import figure, output_notebook, show
from bokeh.models import DatetimeTickFormatter

import json
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

```python
pickles = []
for x in range(0,40):
    filename = "catDate_sentScores/catDate_sentScores_"+str(x)+"k_"+str(x+1)+"k.pickle"
    t = pd.read_pickle(filename)
    pickles.append(t)
t = pd.read_pickle("catDate_sentScores/catDate_sentScores_40k_end.pickle")
pickles.append(t)
df_subcategory_sentiment = pd.concat(pickles)
df_subcategory_sentiment.shape
```

```python
labels = list(set(list(map(lambda x: x[1], df_subcategory_sentiment.index))))
```

```python
# through the whole timeline 
def plotSentiment_subCategory(label):
    idx = list(map(lambda x: x[1] == label, df_subcategory_sentiment.index))
    evolution = df_subcategory_sentiment[idx]
    evolution = evolution.sort_index()
    xaxis = list(map(lambda x: x[0], evolution.index))

    # output to static HTML file
    #output_file("Videos_evolution.html", title="Evolution of the number of reviews through the years")
    output_notebook()
    # create a new plot with a a datetime axis type
    p = figure(width=900, height=500, x_axis_type="datetime", y_range=(-1.1, 1.1), x_range=(852076800000, 1419984000000) )

    # add renderers
    p.line(xaxis,evolution.sentScore, color='navy')
    # NEW: customize by setting attributes
    p.title.text = label
    p.legend.location = "top_left"
    p.grid.grid_line_alpha=0
    p.xaxis.axis_label = 'Date'
    p.yaxis.axis_label = 'Sentiment Score'
    p.yaxis.bounds = [-1,1]
    p.xaxis[0].formatter = DatetimeTickFormatter(days='%m/%d',years='%Y',months='%B')
    # show the results
    show(p)
```

```python
for label in labels[:2]:
    plotSentiment_subCategory(label)
```
