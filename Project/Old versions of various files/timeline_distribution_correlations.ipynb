** Now the daily median of the variables : rating , length of the review , length of the title **

```python
#function to plot the daily mean of a given variable from 1996-2014 
evolution = df.groupby('reviewTime').mean()
evolution.year=list(map(lambda x:int(x) , evolution.year))
del evolution['unixReviewTime']
```

```python
#helpfulness:
time=df.reviewTime.unique()
tab=pd.DataFrame([],index=df.reviewTime.unique())
l=[]
for date in tab.index:
    data=df[df.reviewTime==date].helpful
    helpful=sum(map(lambda x:x[0],data))
    total=sum(map(lambda x:x[1],data))
    if total == 0 :
        l.append(0)
    else:
        l.append(round(helpful/total,2))

tab['helpfulness']=l

evolution=pd.concat([evolution,tab],axis=1)
```

```python
#add count
evolution=pd.concat([evolution,count],axis=1)
```

## b) Monthly average By year :

```python
#function : takes a list of years and plots the monthly mean for the variable chosen above 
def monthly_mean_Y (Y,variable):
    output_notebook()
    # output to static HTML file
    #output_file("Videos_Number.html", title="Number of reviews")
    if all(x in Years for x in Y):
        p = figure(width=1000, height=500, x_axis_type="datetime")
        colors = ['#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99','#e31a1c','#fdbf6f','#ff7f00','#cab2d6','#e41a1c','#377eb8','#4daf4a','#984ea3','#ff7f00','#ffff33','#a65628','#f781bf','#999999']
        k=0
        for y in Y:
            data=monthly_mean[round(monthly_mean.year)==y]
            data.index = data.index.map(lambda t: t.replace(year=2008))
            p.line(data.index,data[variable],color=colors[k],legend=str(y))
            points=evolution[evolution.year==y]
            points.index = points.index.map(lambda t: t.replace(year=2008))
            p.circle(points.index,points[variable],color=colors[k], alpha=0.4,legend=str(y))
            k=k+1
        p.title.text = "monthly mean of "+str(variable)
        p.legend.location = "top_right"
        p.grid.grid_line_alpha=0
        p.xaxis.axis_label = 'Date'
        p.yaxis.axis_label = 'montly mean'
        p.xaxis[0].formatter = DatetimeTickFormatter(days='%m/%d',years='',months='%B')
        return p
    else :
        print('ERROR: Years not in the dataframe')
#show(p)
```

```python
fig=monthly_mean_Y([2006,2007,2012,2013,2014],'length_review')
show(fig)
```

## a) Summary

```python
monthly_avg = r.median()
monthly_avg.describe() 
```

```python
def distribution (variable):
    p1 = figure(width=500, plot_height=500, title=None)
    measured = monthly_avg[variable]
    hist, edges = np.histogram(measured, bins=50)
    x = np.linspace(-max(measured), max(measured), 1000)
    p1.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],fill_color="#e6550d", line_color="#033649")
    p1.legend.location = "center_right"
    p1.legend.background_fill_color = "darkgrey"
    p1.xaxis.axis_label = variable
    p1.yaxis.axis_label = 'count'
    return p1
```

```python
#from bokeh.io import vplot
output_notebook()
score=distribution('overall')
length=distribution('length_review')
helpf=distribution('helpfulness')
number=distribution('asin')
p = gridplot([[number, length],[helpf, score]])

show(p)
```

## b) Correlations between different characteristics of the product :

```python
def corr (x,y,Year):
    p = figure(width=1000, height=500)
    p.circle(evolution.loc[evolution.year>=Year,x],evolution.loc[evolution.year>=Year,y],alpha=0.5,legend='after '+str(Year))
    p.circle(evolution.loc[evolution.year<Year,x],evolution.loc[evolution.year<Year,y],color='green',alpha=0.5,legend='before '+str(Year))
    p.legend.location = "top_left"
    p.grid.grid_line_alpha=0
    p.xaxis.axis_label = x
    p.yaxis.axis_label = y
    
    show(p)
```

```python
corr('overall','helpfulness',2013)
```
